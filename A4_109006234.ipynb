```python
#Question 4
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from datetime import datetime

# ---------- using SVD ----------------
def SVDDistance(spam, ham, test):
    U1 = np.linalg.svd(spam, full_matrices=False)[0]
    X1 = np.dot(U1.T, test)
    R1 = np.dot(U1, X1)-test
    
    U2 = np.linalg.svd(ham, full_matrices=False)[0]
    X2 = np.dot(U2.T, test)
    R2 = np.dot(U2, X2)-test
    
    [m, n] = R1.shape
    pos = 0
    neg = 0
    for i in range(n):
        d1 = np.linalg.norm(R1[:,i]) #distance to the spam subspace
        d2 = np.linalg.norm(R2[:,i]) #distance to the ham subspace
        if d1 >= d2: neg = neg + 1 #not spam
        else: pos = pos + 1 #spam
    return pos, neg

# ---------- main program --------------------
start_time = datetime.now()

emails = pd.read_csv("emails_up.csv", encoding='cp1252')
N = 1143
df = pd.DataFrame(emails)
spam = df['spam']
flod = df['fold']

# initialize
cv = TfidfVectorizer(stop_words='english', max_features=20000, token_pattern=r"(?u)\b[a-zA-Z]\w+\b") 
dt_matrix = cv.fit_transform(df['text']) 
[m, n] = dt_matrix.shape

spam_train = (dt_matrix[[i for i in range(m) if flod[i]<4 and spam[i]==1], :]).toarray().transpose()
spam_test  = (dt_matrix[[i for i in range(m) if flod[i]==4 and spam[i]==1], :]).toarray().transpose()
ham_train = (dt_matrix[[i for i in range(m) if flod[i]<4 and spam[i]==0], :]).toarray().transpose()
ham_test  = (dt_matrix[[i for i in range(m) if flod[i]==4 and spam[i]==0], :]).toarray().transpose()

# Compute the confusion matrix
TP, FN = SVDDistance(spam_train, ham_train, spam_test)
FP, TN = SVDDistance(spam_train, ham_train, ham_test)
print(TP, FN)
print(FP, TN) 

prediction = ((TP+TN)/N)*100
print('Prediction =', prediction)
accuracy = (TP/(TP+FP))*100
print('Accuracy =', accuracy)
recall = (TP/(TP+FN))*100
print('Recall =', recall)

end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```
